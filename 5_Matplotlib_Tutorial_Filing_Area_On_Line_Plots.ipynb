```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## Simple line plot 

```python
data = pd.read_csv('dev_salary_by_age.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.style.use('seaborn')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

## Filling up the line area 

```python
data = pd.read_csv('dev_salary_by_age.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.style.use('seaborn')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287
plt.legend()

# Fill the area
plt.fill_between(ages, py_salaries, alpha = 0.25)
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

## Fill between Python salary and overall median 

```python
data = pd.read_csv('dev_salary_by_age.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.style.use('seaborn')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.legend()

# Fill the area
plt.fill_between(ages, py_salaries, overall_median, alpha = 0.25)
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

## Using different fill colors above and below median salaries

```python
data = pd.read_csv('dev_salary_by_age.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.style.use('seaborn')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.legend()

# Fill the area
plt.fill_between(ages, py_salaries, overall_median, 
                 where = (py_salaries > overall_median),
                 interpolate=True,
                 alpha = 0.25)
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

## Label the filled up section

```python
data = pd.read_csv('dev_salary_by_age.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.style.use('seaborn')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287


# Fill the area
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where = (py_salaries > dev_salaries),
                 interpolate=True,
                 alpha = 0.25, label = 'Above Avg.')
plt.fill_between(ages, py_salaries, overall_median, 
                 where = (py_salaries <= dev_salaries),
                 interpolate=True,color='red',
                 alpha = 0.25, label = 'Below Avg.')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```
