## Using real world data to plot the charts

```python
plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio'] 
plt.scatter(view_count, likes, edgecolors='black', linewidth = 1, alpha = 0.75)


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

## Making the scale logerithmic

```python
plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio'] 
plt.scatter(view_count, likes, edgecolors='black', linewidth = 1, alpha = 0.75)
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

## Using the ratio of likes as color

```python
plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio'] 
plt.scatter(view_count, likes, c=ratio, cmap = 'summer', edgecolors='black', linewidth = 1, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```
